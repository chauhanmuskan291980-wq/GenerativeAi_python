```python
## Temperature Conversion

def connvert_temperature(temp,unit):
    """This function converts temp between Celsius and Faherenheit"""
    if unit=='C':
        return temp * 9/5 + 32
    elif unit=="F":
        return (temp-32)*5/9
    else:
        return None
    
print(connvert_temperature(25,"C"))
print(connvert_temperature(77,"C"))
```

```python
## Password Strength Checker

def is_strong_password(password):
    if len(password)<8:
        return False
    if not any(char.isdigit() for char in password):
        return False
    if not any (char.islower() for char in password):
        return False
    if not any (char.isupper() for char in password):
        return False
    if not any(char in '!@#$%^&*()_+' for char in password):
        return False
    
    return True
    

print(is_strong_password("muskan!!@00$"))
```

```python
def calculate_total_cost(cart):
    total_cost = 0
    for item in cart:
        item_total = float(item['price']) * item['quantity']
        total_cost += item_total
        print(f"{item['name']} : total cost:{item_total}")
    return total_cost

cart = [
    {'name': 'Apple', 'price': '0.5', 'quantity': 4},
    {'name': 'Mango', 'price': '4.5', 'quantity': 7},
    {'name': 'Orange', 'price': '4.5', 'quantity': 2},
    {'name': 'Pears', 'price': '3.5', 'quantity': 5}
]

print("Total:",calculate_total_cost(cart))

   
```

```python
## Check if a string a Palindrome
 
def check_palindromic(string):
    start = 0
    end = len(string) - 1

    while start < end:
        if string[start] != string[end]:
            return False
        start += 1
        end -= 1

    return True

print(check_palindromic("abab"))


def check(string):
    string = string.lower().replace(" ","")
    print(string)
    print(string[::-1])
    return string == string[::-1]

check("abcbacx")




```

```python
# Calculate the factorial of a number by using recursion

def factorial(n):
    if n==0:
        return 1
    else:
        return n* factorial(n-1)
    
factorial(5)
```

```python
# Function to read a file and count the frequency of each word
def count_word_frequency(file_path):
    word_count = {}
    with open(file_path,'r') as file:
        for line in file:
            words = line.split()
            for word in words:
                word = word.lower().strip('.,!?;:"\ ')
                word_count[word] = word_count.get(word,0)+1
    return word_count
filepath='sample.txt'
word_frequency = count_word_frequency(filepath)
print(word_frequency)
    
```
