Lambda functions are small anonymous function defined using lanbda keyword , They cab have any number of arguments but only one expression . They are commonly used for the short operations or as arguments to higher-order functions.
 

```python
def addtion(a,b):
    return a+b

addtion(8,9)
```

```python
addtion = lambda a,b: a+b
print(type(addtion))
addtion(5,6)
```

```python
even = lambda num : num%2==0
even(14)
```

```python
addtion1 = lambda x,y,z : x+y+z
addtion1(8,8,9)
```

```python
## map() - appled a function to all item in a list

numbers=[1,2,3,4,5,6]
def square(number):
    return number**2


list(map(lambda x:x**2 , numbers))


```
