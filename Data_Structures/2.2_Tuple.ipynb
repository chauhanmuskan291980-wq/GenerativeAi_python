Tuples are ordered collections of items that are immutable. They are similar to lists , But their immutability makes them different.


```python
## Create a tuple
empty_tuple = ()
print(type(empty_tuple))
```

```python
lst = list()
print(type(lst))
tpl = tuple()
print(type(tpl))
```

```python
numbers = tuple([1,2,3,4,5,6])
print(numbers)
```

```python
mixed_tuple = (1,"Hello world",1.09,True)
print(mixed_tuple)
```

```python
## Accessing Tuple Elements
print(numbers[0])
print(numbers[1])
print(numbers[0:4])
print(numbers[-1])
```

```python
## Tuple Operations

concatenation_tuple = numbers + mixed_tuple
print(concatenation_tuple)
```

```python
mixed_tuple*3
numbers*3
```

```python
## Immtable Nature of Tuples
## Tuples are immutables , meaning their elements cannot be changed once assigned.

lst =  [1,2,3,5,5]
print(lst)

lst[1] = "Krish"
print(lst)

```

```python
# numbers[1] = "muskan"
# print(numbers)

## Tuple Methods
print(numbers.count(1))
print(numbers.index(3))
# packed tuple
packed_tuple = 1,"hello",3.14
print(packed_tuple)

# Unpacked a tuple
a,b,c = packed_tuple
print(a)
print(b)
print(c)



```

```python
numbers = (1,2,3,4,5,6)
first , *middle , last = numbers
print(first)
print(middle)
print(last)
```

```python
## Netsed Tuple

nested_tuple=((1,2,3),("a","b","c"),(True, False))
nested_tuple[1][2]
```

```python
## Iterating over nested tuples
for sub_tuple in nested_tuple:
    for item in sub_tuple:
        print(item , end=" ")
    print()
```
