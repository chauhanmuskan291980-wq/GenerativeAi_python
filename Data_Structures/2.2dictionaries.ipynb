Dictionaries are unordered collections of items . They are store data in key-value pairs.Keys must be unique and immutable ( e.q: string,numbers,or tuples) , while values can be of any type.


```python
## create Dictionaries
empty_dict = {}
print(type(empty_dict))
```

```python
empty_dict = dict()
empty_list = list()
empty_tuple = tuple()
print(type(empty_dict))
print(type(empty_tuple))
print(type(empty_list))
```

```python
students = {"name":"Muskan" , "age":32, "grade":24}
print(students)
print(type(students))
```

```python
students = {"name":"Muskan" , "age":32, "name":24}
print(students)
```

```python
## Accessing Dictionary Elements
students = {"name":"Muskan" , "age":32, "grade":"A"}

print(students['name'])
print(students['age'])
print(students['grade'])


### Asscessing using get() method
print(students.get("name"))
print(students.get("age"))
print(students.get("grade"))
print(students.get("last_name"))
print(students.get("last_name" , "Not Available"))
```

```python
### Modifying Dictionary Elements
## Dictionary are mutable , so you can add , update and delete elements

print(students)

students["age"] = 33
print(students)
students["address"] = "India"
print(students)
```

```python
del students['grade']
print(students)
```

```python
## DIctionary methonds

keys = students.keys()
print(keys)
value = students.values()
print(value)

items = students.items()
print(items)
```

```python
students_copy = students
print(students)
print(students_copy)
students['name'] = "chauhan"
print(students)
print(students_copy)
```

```python
students_copy1 = students.copy()  ## Shallow copy
students['name'] = "chauhan1"
print(students_copy1)
print(students)
```

```python
### Iterating Over Dictionaaries
## You can use loops to itrtate over dictionatries, keys , values or items


for keys  in students.keys():
    print(keys)



for values in students.values():
    print(values)


for key, value in students.items():
    print(f"{key}:{value}")
```

```python
## Nested Dictionaries

students={
    "student1":{"name":"Muskan", "age":32},
    "student2":{"name":"Peter", "age":23},
    "student3":{"name":"Bhuvan", "age":72}
}

print(students)
```

```python
print(students['student1']['name'])
print(students['student2']['name'])
print(students['student3']['name'])
```

```python
# iterating over nested dictionaries

for students_id , students_info in students.items():
    print(f"{students_id} : {students_info}")
    for key, value in students_info.items():
        print(f"{key}:{value}")
```

```python
## Dictionary Comphrehension 
squareInDict = {x:x**2 for x in range(5)}
print(squareInDict)

squareInList = [x**2 for x in range(5)]
print(squareInList)

squareInTuple = tuple(x**2 for x in range(5))
print(squareInTuple)
```

```python
## Contional dictionary comprehension
evens  = {x:x**2 for x in range(10) if x%2==0}
print(evens)
```

```python
## Practical Examples

## Usea dictionary to count he frequency of elemets in list 

numbers=[1,2,3,4,3,21,4,1,3,5,2,4]

frequency = {}

for number in numbers:
    if number in frequency:
        frequency[number]+=1
    else:
        frequency[number] = 1
print(frequency)

 
```

```python
## Marge two dict 

dict1 = {"a":1,"b":2}
dict2 = {"b":3,"c":4}
marge_dict = {**dict1 , **dict2}
print(marge_dict)
```
