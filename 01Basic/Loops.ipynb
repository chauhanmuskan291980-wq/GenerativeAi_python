```python
## Strings

str = "Muskan Chauhan"

for i in str:
    print(i)
```

```python
# while Loop

count =0
while count<5:
    print(count)
    count= count+1
```

```python
# Loop Control Statements
## break

for i in range(10):
    if i==5:
        break
    print(i)


```

```python
## Continue 
## The continue statement skipd the current iteration and continues with next.

for i in range(10):
    if i%2==0:
        continue
    print(i)
```

```python
## pass
## The pass statement is a null operations ; it does nothing

for i in range(10):
    if i==3:
        print("The number is: ", i)
        pass
    print(i)
```

```python
## Netsted Loops
## a loop insode a loop 

for i in range(3):
    for j in range(2):
        print(f"i:{i} and j:{j}")
```
