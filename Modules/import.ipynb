In Python modules and packages help organize and reuse code. here's a comprehensive guide on how to import them.



```python
import math
math.sqrt(16)
```

```python
from math import sqrt , pi
print(sqrt(16))
print(sqrt(8))
print(pi)
```

```python
import numpy as np
np.array([1,2,3,4,5])
```

```python
from math import *
sqrt(16)
```

```python
from packages.maths import addtions
from packages.subpackages.math import multipy
addtions(9,8)
multipy(9,8)
```
