Abstraction is the concept of hiding the complex implementation details and showing only the necessary features of an object. This helps in reducing programming complexity and effect.

```python
from abc import ABC,abstractmethod
class Vehicle(ABC):
    def drive(self):
        print("The vehicle is used for driving")

    @abstractmethod
    def start_engine(self):
        pass

class Car(Vehicle):
    def start_engine(self):
        print("Car engine is started")


def opreate_vehicle(vehicle):
    vehicle.start_engine()
    vehicle.drive()

car = Car()
opreate_vehicle(car)
```
