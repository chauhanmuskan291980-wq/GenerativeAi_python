Inheritance is fundametal concept in object-Oritend Programmming (OOPS) that allows a class to inherit attributes and methods form another class, This lesson covers single inheritance and multiple inheritance demonstrating how to create and use then in python,.

```python
## Inheritance
## Parent Class
class Car:
    def __init__(self,windows,doors,enginetype):
        self.windows = windows
        self.doors  = doors
        self.enginetype= enginetype

    def drive(self):
        print(f"The person will drive the {self.enginetype} car")

    
car1 = Car(4,5,"patrol")
print(car1.drive())
```

```python
class Tesla(Car):
    def __init__(self, windows, doors, enginetype,is_selfdriveing):
        super().__init__(windows, doors, enginetype)
        self.is_selfdriveing = is_selfdriveing

    def selfdriving(self):
        print(f"Tesla supports self driving: {self.is_selfdriveing}")

Tesla1 = Tesla(4,5,"electric",True)
Tesla1.selfdriving()
Tesla1.drive()
```

```python
class Animal:
    def __init__(self,name):
        self.name= name
    
    def speak(self):
        print('Subclass must implement this method')

class Pet:
      def __init__(self,owner):
        self.owner = owner


## Drived Class
class Dog(Animal,Pet):
    def __init__(self,name,owner):
       Animal.__init__(self,name)
       Pet.__init__(self,owner)

    def speak(self):
        return f"{self.name} say woof"
    

## Crate an object

dog = Dog("Buddy","Krish")
print(dog.speak())
print(f"Owner:{dog.owner}")
```
