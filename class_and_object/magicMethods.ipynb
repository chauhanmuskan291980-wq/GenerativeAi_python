```python
class Person:
    def __init__(self,name,age):
         self.name = name
         self.age = age
    
    def __str__(self):
         return f"{self.name} , {self.age} years old"
    
    def __repr__(self):
          return f"Person(name={self.name} , age = {self.age})"
    
person = Person("Muskan",23)
print(person)
print(repr(person))
```

## Common Operator Overloading Magic Methods
__add__(self,other): Adds two objects using the + operator
__sub__(self,other): Subtracts two objects using the - operator
__mul__(self,other):Multiples two objects using the * operators
__truediv__(self,other):Divides two objects using the / operators
__eq__(self,other):Checks of two objects are equal using the == operator.
__lt__(self,other): Checks if one object is less than another using the < operator.


```python
class Vector:
  
 def __init__(self,x,y):
    self.x =x
    self.y = y

 def __add__(self,other):
   return Vector(self.x+other.x , self.y+other.y)
 def __sub__(self,other):
   return Vector(self.x-other.x , self.y-other.y)
 def __mul__(self,other):
   return Vector(self.x*other.x , self.y*other.y)
 def __eq__(self,other):
   return  self.x == other.x and self.y == other.y
 
 def __repr__(self):
    return f"Vector({self.x} , {self.y})"
 
v1 = Vector(3,4)
v2 = Vector(9,8)

print(v1+v2)
print(v1-v2)
print(v1*v2)
print(v1==v2)



```
