File handling is a crucial part of any programming language . Python provides built-in functions and methods to read from and write to files, both text and binary , The lesson will cover the basic of file handling, includling reading and writing text files and binary files.


```python
## Read a Whole file
with open('example.txt','r') as file:
     content = file.read()
     print(content)
```

```python
## Read a file line by line 
with open('example.txt','r') as file:
    for line in file:
        print(line.strip())
```

```python
## Writing a file

with open('example.txt','w') as file:
    file.write("Hello World\n")
    file.write('This is a new line..')
```

```python
## Write a file without overrriding 

with open('example.txt','a') as file:
    file.write("\nAppend operation taking place!")
```

```python
## Writing a list of lines to a file

lines =['\nFirst line \n ','Second line \n ','Third line \n ','Fouth line \n ']
with open('example.txt','a') as file:
    file.writelines(lines)
```

```python
## Bianry files
data = b'\x00\x01\x02\x03\x04'
with open('example.bin','wb') as file:
    file.write(data)
```

```python

with open('example.bin','rb') as file:
     content = file.read()
     print(content)
```

```python
## Read the content from source text file and write to a destination text file

with open('example.txt','r') as source_file:
    content = source_file.read()


with open('destination.txt','w') as destination_file:
    destination_file.write(content)
```

```python
## Read the file and try to find out the words , charcaters and line 

def count_text_file(file_path):
    with open(file_path,'r') as file:
        lines = file.readlines()
        line_count= len(lines)
        word_count = sum(len(line.split()) for line in lines)
        char_count = sum(len(line) for line in lines)
    return line_count , word_count , char_count

file_path = 'example.txt'
lines, words , character= count_text_file(file_path)

print(f'Lines:{lines} , Words:{words} , Charcaters:{character}')
```
